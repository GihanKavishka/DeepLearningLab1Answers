```python
import numpy as np
import matplotlib.pyplot as plt

# Create a random 100,000x1 array from the exponential distribution
random_array = np.random.exponential(scale=1.0, size=100000)

# Create a histogram for the random array
plt.hist(random_array, density=True, bins=100, histtype="step", color="blue", label="Exponential")

# Generate data for uniform and normal distributions for comparison
uniform_data = np.random.uniform(size=100000)
normal_data = np.random.normal(size=100000)

# Plot histograms for uniform and normal distributions
plt.hist(uniform_data, density=True, bins=100, histtype="step", color="green", label="Uniform")
plt.hist(normal_data, density=True, bins=100, histtype="step", color="red", label="Normal")

# Set plot labels and legend
plt.title("Random distributions")
plt.xlabel("Value")
plt.ylabel("Density")
plt.legend(loc="upper right")

# Show the plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Generate values for X and Y within the specified range
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)

# Calculate Z = X^2 + Y^2
Z = X**2 + Y**2

# Create a 3D plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plot the surface
ax.plot_surface(X, Y, Z)

# Set axis labels
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

# Show the plot
plt.show()

```

```python
import seaborn as sns
import pandas as pd
from matplotlib import pyplot as plt

# Load your dataset into a DataFrame (replace 'your_data.csv' with the actual file path)
df = pd.read_csv('/content/sample_data/online_store_customer_data.csv')

# Select the desired features
selected_features = ['Transaction_date', 'Transaction_ID', 'Gender', 'Age', 'Marital_status', 'State_names','Segment','Employees_status', 'Payment_method','Referal','Amount_spent']

# Calculate Pearson correlation matrix
pearson_corr = df[selected_features].corr(method='pearson')

# Calculate Spearman rank correlation matrix
spearman_corr = df[selected_features].corr(method='spearman')

# Create subplots for heatmaps
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# Plot Pearson correlation heatmap
sns.heatmap(pearson_corr, annot=True, cmap='coolwarm', ax=axes[0])
axes[0].set_title('Pearson Correlation')

# Plot Spearman rank correlation heatmap
sns.heatmap(spearman_corr, annot=True, cmap='coolwarm', ax=axes[1])
axes[1].set_title('Spearman Rank Correlation')

# Show the heatmaps
plt.show()
```

```python
import pandas as pd

# Load the dataset into a DataFrame (replace 'your_data.csv' with the actual file path)
df = pd.read_csv('/content/Au_nanoparticle_dataset.csv')

# Select the desired columns
selected_columns = ['N_total', 'N_bulk', 'N_surface', 'R_avg']

# Create a new DataFrame with only the selected columns
new_df = df[selected_columns]
```

```python
print(new_df.head(20))
```

```python
# Calculate mean for each feature
mean_values = new_df.mean()

# Calculate standard deviation for each feature
std_deviation = new_df.std()

# Calculate quartiles (25%, 50%, 75%) for each feature
quartiles = new_df.quantile([0.25, 0.5, 0.75])

# Display the calculated statistics
print("Mean Values:")
print(mean_values)

print("\nStandard Deviation:")
print(std_deviation)

print("\nQuartiles:")
print(quartiles)
```

```python
import matplotlib.pyplot as plt

# Create a 1x4 layout of histograms
fig, axes = plt.subplots(1, 4, figsize=(16, 4))

# Plot histograms for each feature
for i, column in enumerate(selected_columns):
    axes[i].hist(new_df[column], bins=20, color='blue', alpha=0.7)
    axes[i].set_title(column)
    axes[i].set_xlabel("Value")
    axes[i].set_ylabel("Frequency")

plt.tight_layout()
plt.show()
```

```python
import seaborn as sns

# Visualize scatter plots and histograms using pairplot
sns.pairplot(new_df)
plt.show()
```

```python
# Create a customized pairplot
g = sns.PairGrid(new_df)

# Plots on the diagonal: Histograms along with kernel density estimation plots
g.map_diag(sns.histplot, kde_kws={'color': 'k'})
g.map_offdiag(sns.scatterplot, s=5)

# Plots on the lower half: Bivariate kernel density estimation plots
g.map_lower(sns.kdeplot, cmap='Blues_d')

plt.show()
```
